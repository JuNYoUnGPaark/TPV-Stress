```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 팩트 체크를 위한 가상의 맥파(PPG) 신호 생성
# (기본 심장 박동 파형 + 미세한 혈관 반사파 굴곡 추가)
t = np.linspace(0, 4 * np.pi, 200)
sig = np.sin(t) + 0.3 * np.sin(2 * t + 1.5)

# 2. 질문자님의 코드와 동일한 방식의 지연 임베딩 (1-step delay)
# sig[:-1] 은 현재 값 (X축)
# sig[1:] 은 다음 값 (Y축)
x_coords = sig[:-1]
y_coords = sig[1:]

# 3. 결과 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# [왼쪽] 원래의 1차원 시계열 데이터
axes[0].plot(sig, color='#1f77b4', linewidth=2)
axes[0].set_title("1D Time Series (Original Signal)", fontsize=14)
axes[0].set_xlabel("Time (Samples)")
axes[0].set_ylabel("Amplitude")
axes[0].grid(True, linestyle='--', alpha=0.6)

# [오른쪽] 2차원으로 펼친 위상 공간 (Phase Space)
axes[1].plot(x_coords, y_coords, color='#d62728', linewidth=1.5, marker='o', markersize=3, alpha=0.8)
axes[1].set_title("2D Phase Space (Delay Embedding)", fontsize=14)
axes[1].set_xlabel("x(t) : Current Value")
axes[1].set_ylabel("x(t+1) : Next Value")
axes[1].grid(True, linestyle='--', alpha=0.6)
axes[1].axis('equal') # 비율을 1:1로 맞춰서 찌그러지지 않게 함

plt.tight_layout()
plt.show()
```
