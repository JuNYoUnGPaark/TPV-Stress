```python
import os
import glob
import pickle
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.signal import butter, filtfilt, find_peaks
from scipy.stats import iqr, skew, kurtosis

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

warnings.filterwarnings("ignore")

# =============================================================================
# CONFIG
# =============================================================================
PKL_DIR = r"/content/drive/MyDrive/Colab Notebooks/BP/data/WESAD/pkl_data"

EDA_TPV_CSV = r"/content/drive/MyDrive/Colab Notebooks/BP/TPV-csv/noQC_EDA.csv"
BVP_TPV_CSV = r"/content/drive/MyDrive/Colab Notebooks/BP/TPV-csv/noQC.csv"

OUTPUT_DIR = r"/mnt/data/personalized_eval_results_direct_features"
os.makedirs(OUTPUT_DIR, exist_ok=True)

FS_LABEL = 700
FS_BVP = 64
FS_EDA = 4

WINDOW_SEC = 60
STEP_SEC = 60
MAJ_RATIO_TH = 0.95

LABEL_COL = "label_major"
SUBJECT_COL = "subject"
MERGE_KEYS = ["subject", "label_major", "t_start_sec", "t_end_sec"]

TEST_SIZE = 0.2
RANDOM_STATE = 42

LABEL_BASELINE = 1
LABEL_STRESS = 2
LABEL_MEDITATION = 4
VALID_LABELS = [1, 2, 4]

# filters
BVP_LOWCUT = 0.5
BVP_HIGHCUT = 8.0
EDA_LOWCUT = 0.05
EDA_HIGHCUT = 1.0
FILTER_ORDER = 4

EDA_FEATURE_COLS = [
    "EDA_mean",
    "EDA_std",
    "EDA_min",
    "EDA_max",
    "EDA_median",
    "EDA_iqr",
    "EDA_skew",
    "EDA_kurtosis",
    "EDA_slope",
    "EDA_range",
    "EDA_diff_mean",
    "EDA_diff_std",
    "EDA_num_peaks",
    "EDA_peak_mean",
    "EDA_peak_max",
]

EDA_TPV_COLS = [
    "TPV27", "TPV1", "TPV12", "TPV16", "TPV0", "TPV2", "TPV3", "TPV4",
    "TPV9", "TPV15", "TPV17", "TPV20", "TPV21", "TPV22", "TPV24", "TPV5",
    "TPV6", "TPV8", "TPV10", "TPV11", "TPV23", "TPV28", "TPV30", "TPV31"
]

BVP_FEATURE_COLS = [
    "BVP_mean",
    "BVP_std",
    "BVP_min",
    "BVP_max",
    "BVP_median",
    "BVP_iqr",
    "BVP_skew",
    "BVP_kurtosis",
    "BVP_slope",
    "BVP_range",
    "BVP_diff_mean",
    "BVP_diff_std",
    "BVP_num_peaks",
    "BVP_peak_mean",
    "BVP_peak_max",
]

BVP_TPV_COLS = [
    "TPV22", "TPV9", "TPV4", "TPV2", "TPV0",
    "TPV28", "TPV8", "TPV19", "TPV14", "TPV26", "TPV11"
]


# =============================================================================
# MODEL DEFINITIONS
# =============================================================================
def get_models(num_classes: int):
    objective = "binary:logistic" if num_classes == 2 else "multi:softprob"

    return {
        "XGBoost": XGBClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=5,
            objective=objective,
            random_state=42,
            use_label_encoder=False,
            eval_metric="logloss",
        ),
        "RandomForest": RandomForestClassifier(
            n_estimators=100,
            random_state=42
        ),
        "SVM": SVC(
            kernel="rbf",
            C=1.0,
            probability=True,
            random_state=42
        ),
        "MLP": MLPClassifier(
            hidden_layer_sizes=(64, 32),
            max_iter=500,
            alpha=1e-4,
            solver="adam",
            random_state=42
        )
    }


# =============================================================================
# BASIC UTILS
# =============================================================================
def fill_nan_with_median(x: np.ndarray) -> np.ndarray:
    x = np.asarray(x, dtype=np.float32).reshape(-1).copy()
    if np.isnan(x).any():
        med = np.nanmedian(x)
        if np.isnan(med):
            med = 0.0
        x = np.nan_to_num(x, nan=med, posinf=med, neginf=med)
    return x


def bandpass_filter(sig, fs, low, high, order=4):
    sig = np.asarray(sig, dtype=np.float32).reshape(-1)
    nyq = 0.5 * fs
    low_n = low / nyq
    high_n = high / nyq

    if not (0 < low_n < high_n < 1):
        raise ValueError(f"Invalid bandpass range: low={low}, high={high}, fs={fs}")

    b, a = butter(order, [low_n, high_n], btype="band")
    return filtfilt(b, a, sig).astype(np.float32)


def safe_skew(x: np.ndarray) -> float:
    x = np.asarray(x, dtype=np.float32).reshape(-1)
    if len(x) > 2 and np.std(x) > 1e-12:
        v = skew(x)
        return float(v) if np.isfinite(v) else 0.0
    return 0.0


def safe_kurtosis(x: np.ndarray) -> float:
    x = np.asarray(x, dtype=np.float32).reshape(-1)
    if len(x) > 3 and np.std(x) > 1e-12:
        v = kurtosis(x)
        return float(v) if np.isfinite(v) else 0.0
    return 0.0


def round_merge_keys(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    for c in ["t_start_sec", "t_end_sec"]:
        if c in df.columns:
            df[c] = df[c].round(3)
    return df


def add_prefix(df: pd.DataFrame, cols: list, prefix: str) -> pd.DataFrame:
    df = df.copy()
    rename_map = {c: f"{prefix}_{c}" for c in cols}
    return df.rename(columns=rename_map)


def validate_columns(df: pd.DataFrame, required_cols: list, name: str):
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise ValueError(f"[{name}] Missing columns: {missing}")


# =============================================================================
# WINDOWING HELPERS
# =============================================================================
def find_target_blocks(labels: np.ndarray):
    labels = np.asarray(labels).reshape(-1)
    if len(labels) == 0:
        return {}

    segments = []
    start = 0
    cur = labels[0]
    for i in range(1, len(labels)):
        if labels[i] != cur:
            segments.append((int(cur), start, i))
            start = i
            cur = labels[i]
    segments.append((int(cur), start, len(labels)))

    baseline_seg = None
    stress_seg = None
    meditation_seg = None

    for lab, s, e in segments:
        if lab == LABEL_BASELINE:
            baseline_seg = (s, e)
            break

    for lab, s, e in segments:
        if lab == LABEL_STRESS:
            stress_seg = (s, e)
            break

    for lab, s, e in reversed(segments):
        if lab == LABEL_MEDITATION:
            meditation_seg = (s, e)
            break

    out = {}
    if baseline_seg is not None:
        out[LABEL_BASELINE] = baseline_seg
    if stress_seg is not None:
        out[LABEL_STRESS] = stress_seg
    if meditation_seg is not None:
        out[LABEL_MEDITATION] = meditation_seg
    return out


def window_fully_in_block(start_l: int, end_l: int, label_major: int, target_blocks: dict) -> bool:
    if label_major not in target_blocks:
        return False
    blk_s, blk_e = target_blocks[label_major]
    return (start_l >= blk_s) and (end_l <= blk_e)


# =============================================================================
# DIRECT FEATURE EXTRACTION
# =============================================================================
def compute_basic_signal_features(sig: np.ndarray, prefix: str, fs: int, peak_distance_sec: float, peak_prom_factor: float):
    sig = np.asarray(sig, dtype=np.float32).reshape(-1)

    feats = {}
    feats[f"{prefix}_mean"] = float(np.mean(sig))
    feats[f"{prefix}_std"] = float(np.std(sig))
    feats[f"{prefix}_min"] = float(np.min(sig))
    feats[f"{prefix}_max"] = float(np.max(sig))
    feats[f"{prefix}_median"] = float(np.median(sig))
    feats[f"{prefix}_iqr"] = float(iqr(sig))
    feats[f"{prefix}_skew"] = safe_skew(sig)
    feats[f"{prefix}_kurtosis"] = safe_kurtosis(sig)

    x = np.arange(len(sig), dtype=np.float32)
    if len(sig) > 1:
        slope = np.polyfit(x, sig, 1)[0]
    else:
        slope = 0.0
    feats[f"{prefix}_slope"] = float(slope)

    feats[f"{prefix}_range"] = float(np.max(sig) - np.min(sig))

    if len(sig) > 1:
        diff_sig = np.diff(sig)
        feats[f"{prefix}_diff_mean"] = float(np.mean(diff_sig))
        feats[f"{prefix}_diff_std"] = float(np.std(diff_sig))
    else:
        feats[f"{prefix}_diff_mean"] = 0.0
        feats[f"{prefix}_diff_std"] = 0.0

    distance = max(1, int(round(fs * peak_distance_sec)))
    prominence = max(1e-6, peak_prom_factor * float(np.std(sig)))
    peaks, _ = find_peaks(sig, distance=distance, prominence=prominence)

    feats[f"{prefix}_num_peaks"] = int(len(peaks))

    if len(peaks) > 0:
        peak_vals = sig[peaks]
        feats[f"{prefix}_peak_mean"] = float(np.mean(peak_vals))
        feats[f"{prefix}_peak_max"] = float(np.max(peak_vals))
    else:
        feats[f"{prefix}_peak_mean"] = 0.0
        feats[f"{prefix}_peak_max"] = 0.0

    return feats


def build_direct_feature_tables_from_pkl(pkl_dir: str):
    pkl_list = sorted(glob.glob(os.path.join(pkl_dir, "S*.pkl")))

    eda_rows = []
    bvp_rows = []

    print(f"[INFO] Found {len(pkl_list)} pkl files for direct feature extraction")

    for pkl_path in pkl_list:
        subject_name = os.path.splitext(os.path.basename(pkl_path))[0]
        print(f"[INFO] Processing direct features: {subject_name}")

        with open(pkl_path, "rb") as f:
            s = pickle.load(f, encoding="latin1")

        bvp = np.asarray(s["signal"]["wrist"]["BVP"]).reshape(-1).astype(np.float32)
        eda = np.asarray(s["signal"]["wrist"]["EDA"]).reshape(-1).astype(np.float32)
        labels = np.asarray(s["label"]).reshape(-1).astype(np.int64)

        bvp = fill_nan_with_median(bvp)
        eda = fill_nan_with_median(eda)

        bvp = bandpass_filter(bvp, fs=FS_BVP, low=BVP_LOWCUT, high=BVP_HIGHCUT, order=FILTER_ORDER)
        eda = bandpass_filter(eda, fs=FS_EDA, low=EDA_LOWCUT, high=EDA_HIGHCUT, order=FILTER_ORDER)

        dur_sec = min(len(bvp) / FS_BVP, len(eda) / FS_EDA)
        labels = labels[: int(dur_sec * FS_LABEL)]

        target_blocks = find_target_blocks(labels)
        if len(target_blocks) == 0:
            continue

        Wl = int(WINDOW_SEC * FS_LABEL)
        Sl = int(STEP_SEC * FS_LABEL)
        Wb = int(WINDOW_SEC * FS_BVP)
        We = int(WINDOW_SEC * FS_EDA)

        for start_l in range(0, len(labels) - Wl + 1, Sl):
            end_l = start_l + Wl
            win_lab = labels[start_l:end_l]

            binc = np.bincount(win_lab)
            maj = int(np.argmax(binc))
            maj_ratio = float((win_lab == maj).mean())

            if maj not in VALID_LABELS:
                continue
            if maj_ratio < MAJ_RATIO_TH:
                continue
            if not window_fully_in_block(start_l, end_l, maj, target_blocks):
                continue

            t0 = start_l / FS_LABEL
            t1 = t0 + WINDOW_SEC

            start_b = int(round(t0 * FS_BVP))
            end_b = start_b + Wb
            start_e = int(round(t0 * FS_EDA))
            end_e = start_e + We

            if end_b > len(bvp) or end_e > len(eda):
                continue

            seg_bvp = bvp[start_b:end_b]
            seg_eda = eda[start_e:end_e]

            if len(seg_bvp) != Wb or len(seg_eda) != We:
                continue

            base_info = {
                "subject": subject_name,
                "label_major": maj,
                "t_start_sec": float(t0),
                "t_end_sec": float(t1),
            }

            eda_feats = compute_basic_signal_features(
                seg_eda, prefix="EDA", fs=FS_EDA,
                peak_distance_sec=1.0, peak_prom_factor=0.1
            )
            bvp_feats = compute_basic_signal_features(
                seg_bvp, prefix="BVP", fs=FS_BVP,
                peak_distance_sec=0.4, peak_prom_factor=0.2
            )

            row_eda = base_info.copy()
            row_eda.update(eda_feats)
            eda_rows.append(row_eda)

            row_bvp = base_info.copy()
            row_bvp.update(bvp_feats)
            bvp_rows.append(row_bvp)

    df_eda_feat = pd.DataFrame(eda_rows)
    df_bvp_feat = pd.DataFrame(bvp_rows)

    df_eda_feat = round_merge_keys(df_eda_feat)
    df_bvp_feat = round_merge_keys(df_bvp_feat)

    return df_eda_feat, df_bvp_feat


# =============================================================================
# DATASET BUILD
# =============================================================================
def build_experiment_datasets():
    df_eda_feat, df_bvp_feat = build_direct_feature_tables_from_pkl(PKL_DIR)

    df_eda_tpv = round_merge_keys(pd.read_csv(EDA_TPV_CSV))
    df_bvp_tpv = round_merge_keys(pd.read_csv(BVP_TPV_CSV))

    validate_columns(df_eda_feat, MERGE_KEYS + EDA_FEATURE_COLS, "EDA_FEATURE_DIRECT")
    validate_columns(df_bvp_feat, MERGE_KEYS + BVP_FEATURE_COLS, "BVP_FEATURE_DIRECT")
    validate_columns(df_eda_tpv, MERGE_KEYS + EDA_TPV_COLS, "EDA_TPV")
    validate_columns(df_bvp_tpv, MERGE_KEYS + BVP_TPV_COLS, "BVP_TPV")

    # 1) EDA-Feature extraction
    df1 = df_eda_feat[MERGE_KEYS + EDA_FEATURE_COLS].copy()
    f1 = EDA_FEATURE_COLS.copy()

    # 2) EDA-TPV
    df2 = df_eda_tpv[MERGE_KEYS + EDA_TPV_COLS].copy()
    f2 = EDA_TPV_COLS.copy()

    # 3) BVP-feature extraction
    df3 = df_bvp_feat[MERGE_KEYS + BVP_FEATURE_COLS].copy()
    f3 = BVP_FEATURE_COLS.copy()

    # 4) BVP-TPV
    df4 = df_bvp_tpv[MERGE_KEYS + BVP_TPV_COLS].copy()
    f4 = BVP_TPV_COLS.copy()

    # 5) BVP-TPV + EDA-feature extraction
    df4_pref = add_prefix(df4, f4, "BVP")
    df1_pref = add_prefix(df1, f1, "EDA")
    df5 = pd.merge(df4_pref, df1_pref, on=MERGE_KEYS, how="inner")
    f5 = [f"BVP_{c}" for c in f4] + [f"EDA_{c}" for c in f1]

    # 6) BVP-TPV + EDA-TPV
    df4_pref2 = add_prefix(df4, f4, "BVP")
    df2_pref2 = add_prefix(df2, f2, "EDA")
    df6 = pd.merge(df4_pref2, df2_pref2, on=MERGE_KEYS, how="inner")
    f6 = [f"BVP_{c}" for c in f4] + [f"EDA_{c}" for c in f2]

    experiments = {
        "EDA-Feature extraction": (df1, f1),
        "EDA-TPV": (df2, f2),
        "BVP-Feature extraction": (df3, f3),
        "BVP-TPV": (df4, f4),
        "BVP-TPV + EDA-feature extraction": (df5, f5),
        "BVP-TPV + EDA-TPV": (df6, f6),
    }

    # optional save
    df1.to_csv(os.path.join(OUTPUT_DIR, "EDA_feature_extraction_direct.csv"), index=False)
    df3.to_csv(os.path.join(OUTPUT_DIR, "BVP_feature_extraction_direct.csv"), index=False)

    return experiments


# =============================================================================
# PERSONALIZED EVALUATION
# =============================================================================
def run_personalized_evaluation(
    data: pd.DataFrame,
    feature_cols: list,
    task_name: str,
    class_names: list,
    label_values: list
):
    data = data[data[LABEL_COL].isin(label_values)].copy()

    le = LabelEncoder()
    data["target"] = le.fit_transform(data[LABEL_COL])

    subjects = sorted(data[SUBJECT_COL].unique())
    num_classes = len(class_names)
    models_dict = get_models(num_classes)

    print(f"\n{'='*100}")
    print(f" Task: {task_name}")
    print(f" Subjects   : {len(subjects)}")
    print(f" Samples    : {len(data)}")
    print(f" Features   : {len(feature_cols)}")
    print(f"{'='*100}")

    fig, axes = plt.subplots(1, 4, figsize=(22, 4))
    fig.suptitle(f"Confusion Matrices: {task_name}", fontsize=16)

    results = []

    for idx, (m_name, model) in enumerate(models_dict.items()):
        all_preds = []
        all_trues = []
        used_subjects = 0
        skipped_subjects = []

        for subj in subjects:
            df_sub = data[data[SUBJECT_COL] == subj].copy()

            if df_sub["target"].nunique() < num_classes:
                skipped_subjects.append(subj)
                continue

            if len(df_sub) < max(10, num_classes * 2):
                skipped_subjects.append(subj)
                continue

            try:
                train_df, test_df = train_test_split(
                    df_sub,
                    test_size=TEST_SIZE,
                    random_state=RANDOM_STATE,
                    stratify=df_sub["target"]
                )
            except Exception:
                skipped_subjects.append(subj)
                continue

            X_train = train_df[feature_cols].copy()
            y_train = train_df["target"].copy()
            X_test = test_df[feature_cols].copy()
            y_test = test_df["target"].copy()

            scaler = StandardScaler()
            X_train = scaler.fit_transform(X_train)
            X_test = scaler.transform(X_test)

            model.fit(X_train, y_train)
            preds = model.predict(X_test)

            all_preds.extend(preds)
            all_trues.extend(y_test)
            used_subjects += 1

        if len(all_trues) == 0:
            print(f"[{m_name:12}] No valid subject split available.")
            continue

        acc = accuracy_score(all_trues, all_preds)
        f1 = f1_score(all_trues, all_preds, average="macro")

        print(f"[{m_name:12}] Acc: {acc:.4f} | F1-Macro: {f1:.4f} | Used: {used_subjects} | Skipped: {len(skipped_subjects)}")
        print(classification_report(all_trues, all_preds, target_names=class_names, digits=4))

        results.append({
            "task": task_name,
            "model": m_name,
            "n_features": len(feature_cols),
            "n_samples": len(data),
            "used_subjects": used_subjects,
            "skipped_subjects": len(skipped_subjects),
            "acc": acc,
            "f1_macro": f1
        })

        cm = confusion_matrix(all_trues, all_preds)
        sns.heatmap(
            cm,
            annot=True,
            fmt="d",
            cmap="Blues",
            ax=axes[idx],
            cbar=False,
            xticklabels=class_names,
            yticklabels=class_names
        )
        axes[idx].set_title(m_name)
        axes[idx].set_xlabel("Predicted")
        axes[idx].set_ylabel("Actual")

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    return pd.DataFrame(results)


# =============================================================================
# MAIN
# =============================================================================
if __name__ == "__main__":
    experiments = build_experiment_datasets()

    all_results = []

    for exp_name, (df_exp, feat_cols) in experiments.items():
        print(f"\n[INFO] Running experiment: {exp_name}")
        print(f"[INFO] Shape: {df_exp.shape}")
        print(f"[INFO] #Features: {len(feat_cols)}")

        print("\n[INFO] Label counts")
        print(df_exp[LABEL_COL].value_counts().sort_index())

        result_bin = run_personalized_evaluation(
            data=df_exp,
            feature_cols=feat_cols,
            task_name=f"{exp_name} | Personalized 80:20 | Binary (Baseline vs Stress)",
            class_names=["Baseline", "Stress"],
            label_values=[1, 2]
        )

        result_tri = run_personalized_evaluation(
            data=df_exp,
            feature_cols=feat_cols,
            task_name=f"{exp_name} | Personalized 80:20 | 3-Class (Baseline vs Stress vs Meditation)",
            class_names=["Baseline", "Stress", "Meditation"],
            label_values=[1, 2, 4]
        )

        all_results.append(result_bin)
        all_results.append(result_tri)

    df_results = pd.concat(all_results, axis=0, ignore_index=True)

    results_csv = os.path.join(OUTPUT_DIR, "personalized_80_20_results_all_versions.csv")
    df_results.to_csv(results_csv, index=False)

    print(f"\n[INFO] Saved results: {results_csv}")
    print("\n[INFO] Final summary")
    print(df_results.to_string(index=False))
```
